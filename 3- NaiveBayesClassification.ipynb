```python
#Dataset ---------->

data = pd.read_csv('01_District_wise_crimes_committed_IPC_2001_2012.csv')
data
```

```python
# Calculating Total Violent Crimes ------------->

data['Total Violent Crimes'] = 0
data['Total Violent Crimes'] = data.iloc[:,2:].sum(axis=1)
data
```

```python
# Convert categorical '# Convert categorical 'STATE/UT' column to one-hot encoding
df = pd.get_dummies(data, columns=['STATE/UT'])


# Create a target variable -------->

df['Crime_Label'] = pd.cut(df['Total Violent Crimes'], bins=3, labels=['Low', 'Medium', 'High'])
df = df.drop(columns=['Total Violent Crimes'])
```

```python
# Prepare features and labels ----------->

X = df.drop(columns=['Crime_Label'])
y = df['Crime_Label']

X,y
```

```python
# Split the data ------------------->

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# train the GaussianNaiveBayes ------------->

# gnb = GaussianNaiveBayes()
# gnb.fit(X_train, y_train)

# # Make predictions -------------->
# y_pred = gnb.predict(X_test)

# # Print predictions and evaluate -------------------->
# print("Predictions:", y_pred)
# print("Accuracy Score:", accuracy_score(y_test, y_pred))


# Fitting Gaussian Naive Bayes model------------>
classes = np.unique(y_train)
means = {}
variances = {}
priors = {}

for cls in classes:
    X_cls = X_train[y_train == cls]
    means[cls] = X_cls.mean(axis=0)
    variances[cls] = X_cls.var(axis=0)
    priors[cls] = len(X_cls) / len(X_train)

def calculate_likelihood(X, cls, means, variances):
    mean = means[cls]
    var = variances[cls]
    likelihood = np.exp(- (X - mean) ** 2 / (2 * var)) / np.sqrt(2 * np.pi * var)
    return np.prod(likelihood, axis=1)

def predict(X, classes, means, variances, priors):
    predictions = []
    for _, row in X.iterrows():
        posteriors = []
        for cls in classes:
            prior = np.log(priors[cls])
            likelihood = np.log(calculate_likelihood(pd.DataFrame([row], columns=X.columns), cls, means, variances))
            posterior = prior + np.sum(likelihood)
            posteriors.append(posterior)
        predictions.append(classes[np.argmax(posteriors)])
    return np.array(predictions)

# Making predictions ------------>
y_pred = predict(X_test, classes, means, variances, priors)

# Print prediction and evaluate --------------->>
print("Predictions:", y_pred)
print("Accuracy Score:", accuracy_score(y_test, y_pred))  # Evaluating the model
```
